# Import data
In this statistical analysis we will use the balanced sample. After one initial analysis we came to the conclusion that the statistical results are similar to the data before the balance

```python
# import the stored data
%store -r data_resampled

print (data_resampled.shape)

print (data_resampled.columns)

print (data_resampled['is_sofw_dev'].value_counts())
```

```python
data_balanced = data_resampled
```

### Attended bootcamps

```python
plt.hist(data_balanced.attended_bootcamp, bins=2)
plt.savefig('/home/inrx/Ironhack/Projects-Ironhack/Machine-Learning-Project/Outputs/attended_bootcamp.png', dpi=480)

print (f"{round(data_balanced['attended_bootcamp'].value_counts()[0]/data_balanced.shape[0]*100)}% didn't attended a bootcamp")

print (f"{round(data_balanced['attended_bootcamp'].value_counts()[1]/data_balanced.shape[0]*100)}% attend a bootcamp.")
```

### Male and Female division

```python
female = data_balanced.gender_female.sum()

male = data_balanced.gender_male.sum()

female_percentage = female/data_balanced.shape[0]*100
male_percentage = male/data_balanced.shape[0]*100

print (f'There are {round(female_percentage)}% female responders to this survey.')

plt.bar(['Female', 'Male'], [female_percentage, male_percentage])
plt.savefig('/home/inrx/Ironhack/Projects-Ironhack/Machine-Learning-Project/Outputs/male_female.png', dpi=480)
```

### Age

```python
sns.distplot(data_balanced.age)
plt.savefig('/home/inrx/Ironhack/Projects-Ironhack/Machine-Learning-Project/Outputs/age.png', dpi=480)

print (f"The average age of people in our sample is {round(data_balanced['age'].mean())}.")
```

### Hours invested in learning Code

```python
sns.distplot(data_balanced.hours_learning_week)
plt.savefig('/home/inrx/Ironhack/Projects-Ironhack/Machine-Learning-Project/Outputs/hours_learning.png', dpi=480)

mean_hours = data_balanced['hours_learning_week'].mean()

print (f'In average, people of our sample spend {round(mean_hours)} hours learning code')
```

### Coding Experience

```python
sns.distplot(data_balanced.months_programming)
plt.savefig('/home/inrx/Ironhack/Projects-Ironhack/Machine-Learning-Project/Outputs/months_programming.png', dpi=480)

mode_experience = mode(data_balanced.months_programming, axis=None)

mean_experience = data_balanced['months_programming'].mean()

print (f'In average, people in the sample have {round(mean_experience)} months of coding experience\
Most of respondents have {mode_experience[0]} months of experience.')

less_experience = round(len(data_balanced[data_balanced['months_programming']<12])/data_balanced.shape[0]*100)

print (f"{less_experience}% of the people in the sample have less than 19 months of experience")

less_3_experience = round(len(data_balanced[data_balanced['months_programming']<3])/data_balanced.shape[0]*100)

print (f"{less_3_experience}% of the sample have less then 3 months of experience in programming")
```

### Results of attending a Bootcamp

```python
attended = data_balanced[data_balanced['attended_bootcamp']==1]
not_sd = round(attended['is_sofw_dev'].value_counts()[0]/attended.shape[0]*100)
sd = round(attended['is_sofw_dev'].value_counts()[1]/attended.shape[0]*100,2)

print (f"{sd}% of the people who attended a Bootcamp (from our sample) got a software developer job")

sns.distplot(attended.is_sofw_dev)
plt.savefig('/home/inrx/Ironhack/Projects-Ironhack/Machine-Learning-Project/Outputs/after_bootcamp', dpi=480)
```
