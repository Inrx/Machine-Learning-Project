# Functions

```python
def to_numeric(lst, df):
    for i in lst:
        df[i] = pd.to_numeric(df[i], errors='coerce')
        df[i] = df[i].fillna(1)
    return df[i]
```

```python
def extra_resources(lst, df, name_column):
    """Create a summary column for the extra resources asked in the survey"""
    df[lst] = df[lst].fillna(0)
    to_numeric(lst=lst, df=df)
    df[name_column] = df.apply(lambda row: row[lst].sum(),axis=1)
    return df[lst]
```

```python
def merge_dataframe (objs, key):
    """Merge DataFrames, given the dfs (objs) and the key (where df will merge)"""
    return reduce(lambda left,right: pd.merge(left,right,on=key), objs)
```
